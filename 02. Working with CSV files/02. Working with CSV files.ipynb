```python
# import pandas for data handling
import pandas as pd
```

## **Reading the CSV file**

```python
# read the file from current folder
df = pd.read_csv('aug_train.csv')

# show the first five rows of the dataset
df.head()
```

```python
# read the file from URL
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

# display first five rows
df.head()
```

## **Using Separator (sep Parameter)**

```python
df = pd.read_csv('movie_titles_metadata.tsv', sep='\t')
df.head()
```

## **Setting Column Names Manually**

```python
df = pd.read_csv('movie_titles_metadata.tsv', sep='\t', names=['sno', 'name', 'release_year', 'rating', 'votes', 'genres'])
df.head()
```

## **Setting an Index Column**

```python
df = pd.read_csv('aug_train.csv', index_col='enrollee_id')
df.head()
```

## **Header Parameter**

```python
df = pd.read_csv('test.csv')
df.head()
```

```python
# if the actual column name starts from row 2 (index 1) we can skip the first row using header=1
df = pd.read_csv('test.csv', header=1)
df.head()
```

## **Selecting the specific columns (Usecols)**

```python
# only load the selected columns instead of full dataset
df = pd.read_csv('aug_train.csv', usecols=['enrollee_id', 'gender', 'education_level'])
df.head()
```

## **Squeeze (Convert Dataframes - Series)**

```python
# when reading the single column, Pandas still returns a Dataframe, squeeze() converts into a series (1D data)
df = pd.read_csv('aug_train.csv', usecols=['enrollee_id'])
enrollee_series = df.squeeze()
enrollee_series
```

## **Skipping Rows (skiprows)**

```python
# skips specific rows while reading the file, here we skip row 0 and 1
df = pd.read_csv('aug_train.csv', skiprows=[0, 1])
df.head()
```

## **Reading only N Rows (nrows)**

```python
# loads only first 100 rows from the file
df = pd.read_csv('aug_train.csv', nrows=100)
df
```

## **Encoding Parameter**

```python
 # some csv files use different text encodings, if we get a UniqueDecodeError, try changing encoding
pd.read_csv('zomato.csv')
```

```python
pd.read_csv('zomato.csv', encoding='latin-1')
```

## **Error Bad Lines (Skip corrupt rows)**

```python
# if some rows are broken or badly formatted, we can skip them using on_bad_lines='skip'
pd.read_csv('BX-Books.csv', encoding="latin-1", on_bad_lines='skip')
```

## **dtypes Parameter**

```python
# dtypes is used to force specific data types for columns
# Useful when : pandas guesses wrong data type, we want to reduce memory usage, we want consistency in column types
pd.read_csv('aug_train.csv').info()
```

```python
pd.read_csv('aug_train.csv', dtype={'target':int}).info()
```

## **Handling dates**

```python
pd.read_csv('IPL Matches 2008-2020.csv').info()
```

```python
# converting a string column into a datetime object for date calculations
pd.read_csv('IPL Matches 2008-2020.csv', parse_dates=['date']).info()
```

## **Convertors**

```python
# applying a custom function to a column during import
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    return name
```

```python
rename('Royal Challengers Bangalore')
```

```python
rename('Kolkata Knight Riders')
```

## **NA Values**

```python
# specifying custom strings that should be treated as NaN (missing values)
pd.read_csv('aug_train.csv', na_values=['Male'])
```

## **Handling Massive Datasets (Chunking)**

```python
dfs = pd.read_csv('aug_train.csv', chunksize=5000)

for df in dfs:
  print(df.shape)
```
