```python
# import pandas for data handling and manipulation
import pandas as pd

# import ProfileReport from ydata-profiling for automated EDA
from ydata_profiling import ProfileReport
```

```python
# read the Titanic training dataset from CSV into a pandas DataFrame
df = pd.read_csv('train.csv')
```

```python
# create a detailed EDA profile report for the Titanic dataset
# explorative=True enables detailed analysis including correlations, missing data, and statistics
profile = ProfileReport(df, title="Titanic Dataset Report", explorative=True)

# display the report interactively within the notebook using an iframe
profile.to_notebook_iframe()
```

```python
# save the EDA profile report as an HTML file for offline viewing
profile.to_file("Titanic_report.html")
```
