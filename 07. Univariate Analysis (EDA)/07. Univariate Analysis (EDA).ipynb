```python
# import pandas for data analysis and working with tables
import pandas as pd

# import seaborn for creating attractive and easy statistical graphs
import seaborn as sns

# import matplotlib for basic plotting and visualization
import matplotlib.pyplot as plt
```

```python
# famous titanic dataset
df = pd.read_csv("./train.csv")
```

```python
df.head()
```

```python
df.info()
```

## **1. Categorical Data**

### **a. Countplot**

```python
# create a bar chart showing the count of values in the "Survived" column
sns.countplot(data=df, x="Survived")
```

```python
# count how many times each value appears in the "Survived" column
df["Survived"].value_counts()
```

```python
# create a bar chart showing passenger class-wise counts, separated by survival status
# (hue is used to add a second categorical variable and show its distribution using different colors)
sns.countplot(data=df, x="Pclass", hue="Survived")
```

```python
# create a bar chart showing gender-wise counts, separated by survival status
sns.countplot(data=df, x="Sex", hue="Survived")
```

```python
# create a bar chart showing the count of passengers for each embarkation port
sns.countplot(data=df, x="Embarked")
```

### **b. Piechart**

```python
# create a pie chart showing the percentage distribution of passenger classes
df["Pclass"].value_counts().plot(kind="pie", autopct="%.2f")
```

## **2. Numerical Data**

### **a. Histogram**

```python
# create a histogram to show the distribution of passengers' ages
plt.hist(df["Age"], bins=8)
```

### **b. Distplot**

```python
# create a histogram of Age showing probability distribution and add a smooth density curve (KDE)
sns.histplot(df, x="Age", kde=True, stat="probability")
```

### **c. Boxplot**

```python
# create a box plot to show the spread and outliers of Age data
sns.boxplot(df, x="Age")
```

```python
# display summary statistics (count, mean, min, max, etc.) for the Age and Fare columns only
df.describe()[["Age", "Fare"]]
```

```python
# calculate the skewness of the Age distribution
df.Age.skew()
```
