```python
# importing pandas for data manipulation
import pandas as pd

# importing request to make API calls
import requests
```

```python
# create an empty DataFrame with the required columns
df = pd.DataFrame()
```

```python
# looping through the pages
for i in range (1, 429):
  # Dynamic URL: The {i} allows the loop to change the page number in each iteration
  url = 'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i)

  # fetching data from the URL
  response = requests.get(url)

  # converting the response to JSON
  data = response.json()

  # extracting 'results' list which contains movie details
  temp_df = pd.DataFrame(data['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]

  # appending the current page's to the main DataFrame
  # ignore_index=True ensures the index contains linearly (0, 1, 2, 3...)
  df = pd.concat([temp_df], ignore_index=True)

# checking the final shape of the data
print("Final Data Shape:", df.shape)

# exporting the final DataFrame to a CSV file
df.to_csv('movies_data.csv', index=False)
```
